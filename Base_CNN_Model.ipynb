```
%tensorflow_version 2.x

import numpy as np
import pandas as pd
import matplotlib as mp
import tensorflow as tf
from tensorflow import keras
```

```
from sklearn.datasets import load_sample_image

#Load sample images 
china = load_sample_image("china.jpg") /255

flower = load_sample_image("flower.jpg")/255

images = np.array([china,flower])
batch_size,height,width,channels = images.shape
```

## How to create Vertical and Horizonatal filters

filters = np.zeros(shape=(7,7,channels,2) , dtype=np.float32)
>**filters[:,3,:,0] = 1** #Vertical filter

>**filters[3,:,:,1] = 1** #Horizonatl Filter

```
#Creating 2 filters

filters = np.zeros(shape=(7,7,channels,2) , dtype=np.float32)
```

```
filters[:,3,:,0] = 1 #Vertical filter
filters[3,:,:,1] = 1 #Horizonatl Filter
```


## Conv2D layers parameters: 

> images = Input mini_batch -A 4D tensor

> filters = Set of filters to apply.

>stride = The sliding of convolution filter on the input layer. Here its set to 1 but it could also be a 1D array with 4 elements. 

>PADDING = **VALID/SAME** 

> VALID = Does not use zero padding and ignores the bottom and right input of the image left out of stride. 

> SAME = Uses zero padding if necessary. No of OUTPUT neurons = No of INPUT Neurons / round(stride). 



```
#Creating a simple Convulation 2D layers

output = tf.nn.conv2d(images,filters,strides=1,padding="SAME")
```

```
#PLotting 1st image feature map
import matplotlib.pyplot as plt

plt.imshow(output[0,:,:,1], cmap="gray")
plt.show()
```
